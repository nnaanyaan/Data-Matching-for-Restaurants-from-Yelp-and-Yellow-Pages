```python
import magellan as mg
```

```python
# initialize JVM
mg.init_jvm()
```

```python
# read in csv files
A= mg.read_csv('tableA.csv',key='id')
B= mg.read_csv('tableB.csv',key='id')
```

```python
# get attribute list
attr_A= list(A.get_attr_names())
attr_B= list(B.get_attr_names())
print attr_A
print attr_B
```

# Include streetAddress in table C

```python
# read in candidate set C
C= mg.read_csv('tableC.csv',ltable= A, rtable= B)
```

```python
# initialize Attribute equivalence blocker
ab = mg.AttrEquivalenceBlocker()
```

```python
# create dummy table (streetAddress included) by selecting different attributes
D= ab.block_tables(A,B,attr_A[1],attr_B[2],
                   l_output_attrs=[attr_A[2]],
                   r_output_attrs=[attr_B[2]])
D
```

```python
import time as t
start= t.time()
print start
```

```python
E= mg.combine_block_outputs_via_union([C,D])
end= t.time()
diff= end-start
print diff
```

```python
E
```

```python
E.to_csv('table_C.csv')
```

# Do further blocking

```python
# read in candidate set C
C= mg.read_csv('table_C.csv',ltable= A, rtable= B)
```

```python
# do another blocking
# do overlap blocking
ob = mg.OverlapBlocker()
```

```python
# overlap on 'name' (qgram=5)
D= ob.block_candset(C,l_overlap_attr=attr_A[1],r_overlap_attr=attr_B[1],qgram=5)
D
```

```python
# overlap on 'streetAddress' (qgram=5,overlap_size=2)
E= ob.block_candset(D,l_overlap_attr=attr_A[2],r_overlap_attr=attr_B[2],qgram=5,overlap_size=2)
E
```

# Do sampling

```python
# now randomly sample 400 instances
S = mg.sample_table(E, 400)
S
```

```python
# output the sample csv
S.to_csv('sample1.csv')
```

```python
# Label S
G = mg.label_table(S, 'gold')
# output G
G.to_csv('OutputG1.csv')
```
