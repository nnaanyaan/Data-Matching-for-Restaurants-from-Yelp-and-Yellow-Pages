```python
import magellan as mg
import numpy as np
```

```python
# read in csv files
A= mg.read_csv('tableA.csv',key='id')
B= mg.read_csv('tableB.csv',key='id')
```

```python
# get attribute list
attr_A= list(A.get_attr_names())
attr_B= list(B.get_attr_names())
```

```python
attr_A
```

```python
attr_B
```

```python
# partition table into 2 based on whether attributes have NaN value
# return 2 MTables in mg.MTable format
def partitionTable(table,attr):
    if type(table[attr][0]) != str:
        t1= table[~np.isnan(table[attr])]
        t2= table[np.isnan(table[attr])]
    else:
        t1= table[table[attr]!='']
        t2= table[table[attr]=='']        
    t1= mg.MTable(t1,key='id')
    t2= mg.MTable(t2,key='id')
    return t1, t2
```

```python
# initialize Attribute equivalence blocker
ab = mg.AttrEquivalenceBlocker()
```

```python
# Attribute equivalence blocking: block A and B using city
C = ab.block_tables(A, B, attr_A[3], attr_B[3], 
                     l_output_attrs=['name'], 
                     r_output_attrs=['name'])
```

```python
C
```

```python
# Attribute equivalence blocking: block A and B using streetAddress
D = ab.block_tables(A, B, attr_A[2], attr_B[2], 
                     l_output_attrs=['name'], 
                     r_output_attrs=['name'])
```

```python
D
```

```python
# initialize JVM
mg.init_jvm()
```

```python
feature_table= mg.get_features_for_blocking(A, B)
```

```python
feature_table
```

```python
# Create rule-based blocker
rb = mg.RuleBasedBlocker()
# Add rule : block tuples if name_name_lev(ltuple, rtuple) < 0.2
rb.add_rule(['name_name_jac_qgm_3_qgm_3(ltuple, rtuple) < 0.2'], feature_table)
```

```python
E = rb.block_candset(C)
```

```python
E
```

```python
# Create another rule-based blocker
rb2 = mg.RuleBasedBlocker()
# Add rule : block tuples if name_name_lev(ltuple, rtuple) >= 0.2
rb2.add_rule(['name_name_jac_qgm_3_qgm_3(ltuple, rtuple) >= 0.2'], feature_table)
```

```python
F = rb2.block_candset(D)
```

```python
F
```

```python
# combine blockers via union
G= mg.combine_block_outputs_via_union([E,F])
```

```python
G
```

```python
# use the debugger
dbg= mg.debug_blocker(A,B,G,output_size= 1000)
```
